# Accuracy Test for Limber Integral

In this notebook we test the accuracy of the limber integration for varying numbers of redshift points.

```python
# Run zIntegral.py to calculate the relevant quantities
# !python zIntegral.py 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Comment out the following if you have not installed niceplots
import niceplots as nicepl
nicepl.initPlot()
```

```python
fn_bispec_kgg="accuracy_files/z_bispec_kgg.dat"
fn_bispec_kkg="accuracy_files/z_bispec_kkg.dat"
```

```python
kgg=np.loadtxt(fn_bispec_kgg)
kkg=np.loadtxt(fn_bispec_kkg)

Ns=[10,20,30,50,100,200]
```

```python
mask=[(not np.isnan(k)) and (k>0) for k in kgg[-1]]

gs_kw=dict(height_ratios=[3,1]) # This sets the ratios between the heights of plots


fig, axes = plt.subplots(nrows=2, sharex=True, gridspec_kw=gs_kw)
plt.subplots_adjust(hspace=0)

for i, N in enumerate(Ns):
    axes[0].plot(kgg[i][mask], label=f'{N} redshift points')
axes[0].legend(ncols=2)

for i, N in enumerate(Ns[:-1]):
    axes[1].plot((kgg[i]/kgg[-1]-1)[mask], label=f'{N}, redshift points')

axes[1].axhspan(-0.1, 0.1, color='k', alpha=0.1)
axes[1].axhspan(-0.05, 0.05, color='k', alpha=0.1)
axes[1].axhline(0, color='k', ls=':')

axes[0].set_yscale('log')
axes[0].set_ylabel(r'$B_{\kappa g g}(\ell_1, \ell_2, \phi)$')
axes[1].set_ylabel("Rel. Diff. \n to finest binning")
axes[1].set_xlabel(r"$(\ell_1, \ell_2, \phi)$ combination")
axes[1].set_ylim(-0.6, 0.6)

nicepl.finalizePlot(axes, tightlayout=False, showlegend=False, outputFn="accuracy_plots/z_integration_Bkgg.pdf")
```

```python
mask=[(not np.isnan(k)) and (k>0) for k in kkg[-1]]

gs_kw=dict(height_ratios=[3,1]) # This sets the ratios between the heights of plots


fig, axes = plt.subplots(nrows=2, sharex=True, gridspec_kw=gs_kw)
plt.subplots_adjust(hspace=0)

for i, N in enumerate(Ns):
    axes[0].plot(kkg[i][mask], label=f'{N} redshift points')
axes[0].legend(ncols=2)

for i, N in enumerate(Ns[:-1]):
    axes[1].plot((kkg[i]/kkg[-1]-1)[mask], label=f'{N}, redshift points')

axes[1].axhspan(-0.1, 0.1, color='k', alpha=0.1)
axes[1].axhspan(-0.05, 0.05, color='k', alpha=0.1)
axes[1].axhline(0, color='k', ls=':')

axes[0].set_yscale('log')
axes[0].set_ylabel(r'$B_{\kappa \kappa g}(\ell_1, \ell_2, \phi)$')
axes[1].set_ylabel("Rel. Diff. \n to finest binning")
axes[1].set_xlabel(r"$(\ell_1, \ell_2, \phi)$ combination")
axes[1].set_ylim(-0.6, 0.6)

nicepl.finalizePlot(axes, tightlayout=False, showlegend=False, outputFn="accuracy_plots/z_integration_Bkkg.pdf")
```
