# Accuracy Test for Aperture statistics integral

In this notebook we test the accuracy of the bispectrum integration for varying numbers of $\ell$ and $\phi$.

```python
# Run zIntegral.py to calculate the relevant quantities
# !python ellPhiIntegral.py 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Comment out the following if you have not installed niceplots
import niceplots as nicepl
nicepl.initPlot()
```

```python
Ns=[5, 10, 20, 30, 50]
```

```python

MNN=[]
for n in Ns:
    MNN.append(np.loadtxt(f"./accuracy_files/MNN_N_{n}.dat")[:,1])
    thetas=np.loadtxt(f"./accuracy_files/MNN_N_{n}.dat")[:,0]
```

```python
gs_kw=dict(height_ratios=[3,1]) # This sets the ratios between the heights of plots


fig, axes = plt.subplots(nrows=2, sharex=True, gridspec_kw=gs_kw)
plt.subplots_adjust(hspace=0)

for i, N in enumerate(Ns):
    axes[0].plot(thetas, MNN[i], label=f'{N} integration points')
axes[0].legend(ncols=1)

for i, N in enumerate(Ns[:-1]):
    axes[1].plot(thetas, (MNN[i]/MNN[-1]-1),label=f'{N}, integration points')

axes[1].axhspan(-0.1, 0.1, color='k', alpha=0.1)
axes[1].axhspan(-0.05, 0.05, color='k', alpha=0.1)
axes[1].axhline(0, color='k', ls=':')

axes[0].set_yscale('log')
axes[1].set_xscale('log')

axes[0].set_ylabel(r'$\langle M_\mathrm{ap}\, N_\mathrm{ap}\, N_\mathrm{ap}\rangle (\theta, \theta, \theta)$')
axes[1].set_ylabel("Rel. Diff. \n to finest binning")
axes[1].set_xlabel(r"$\theta$ [arcmin]")
axes[1].set_ylim(-0.3, 0.3)

nicepl.finalizePlot(axes, tightlayout=False, showlegend=False, outputFn="accuracy_plots/MNN.pdf")
```
